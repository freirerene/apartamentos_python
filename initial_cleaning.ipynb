```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('apartamentos.csv')
```

## Cleaning

First we drop the index column which was carried over from `collect_data_realestate.py` because we didn't use `index=False` as an argument of `to_csv`.

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df
```

#### aluguel
Let's begin cleaning the `aluguel` column. We want it to be a numeric column. At first sight it looks like it contains prices in the form `number` `dot` `number`. Let's see:

```python
pattern_aluguel = r"([0-9]+\.[0-9]+)"
alu = df['aluguel'].str.contains(pattern_aluguel)
```

```python
alu.value_counts()
```

So, actually, `523` rows do not contain that format. What are they, then?

```python
df['aluguel'][~alu].unique()
```

Ok, so they are prices below 1000. So the decimal point might or might not be in the pattern we want.

```python
pattern_aluguel2 = r"([0-9]+?\.?[0-9]+)"
alu2 = df['aluguel'].str.contains(pattern_aluguel2)
alu2.value_counts()
```

We caught all possible values of the rent, then. We used capture groups above so that we can use pattern_aluguel2 to extract the string we want:

```python
df['aluguel'] = df['aluguel'].str.extract(pattern_aluguel2)
```

```python
df['aluguel'] = df['aluguel'].str.replace('.','').astype(float)
```

```python
df['aluguel']
```

#### condominio
Now we do the same thing for `condominio`

```python
df['condominio'].unique()
```

```python
pattern_condominio = r"([0-9]?\.?[0-9]+)"
df['condominio'].str.contains(pattern_condominio).value_counts()
```

```python
df['condominio'] = df['condominio'].str.extract(pattern_condominio)
df['condominio'] = df['condominio'].str.replace('.','').astype(float)
```

```python
df['condominio']
```

#### area

```python
df['area'].unique()
```

```python
pattern_area = r"([0-9]?\.?[0-9]+)"
df['area'].str.contains(pattern_area).value_counts()
```

```python
df['area'] = df['area'].str.extract(pattern_area)
df['area'] = df['area'].str.replace('.','').astype(float)
df['area']
```

#### bairro

```python
df['bairro'] = df['bairro'].str.split(',').str[0].str.strip()
```

```python
df['bairro'].unique()
```

#### quartos, banheiros e garagem

```python
df['quartos'].unique()
```

```python
pattern = r"([0-9])"
df['quartos'].str.contains(pattern).value_counts()
```

```python
df['quartos'] = df['quartos'].str.extract(pattern)
df['quartos'] = df['quartos'].astype(float)
df['quartos']
```

```python
df['banheiros'].unique()
```

```python
df['banheiros'] = df['banheiros'].str.extract(pattern)
df['banheiros'] = df['banheiros'].astype(float)
df['banheiros']
```

```python
df['garagem'].unique()
```

```python
df['garagem'] = df['garagem'].str.extract(pattern)
df['garagem'] = df['garagem'].astype(float)
df['garagem']
```

#### anunciante

```python
df['anunciante'].unique()
```

```python
pattern_anunciante = r"(?<=/imobiliarias/)(.+)(?=_)"
```

## Dealing with the `NaN`s

Now that we cleaned our data its time to deal with the missing values.

```python
df.isnull().sum()
```

```python
df['garagem'].unique()
```

We see the column with the most missing values is the garage column. Inspecting its possible values (i.e., the unique values) we see there's no zero in there; but we know some flats for renting don't have any garages.

We also see when there's no garage available the site leaves the spot empty. So the most reasonable Ansatz to fill the missing values here is to fill it with zeros.

```python
df['garagem'].fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df[df['area'].isnull()]
```

One of the key information when renting an apartament is the area it has. Missing area cannot be replaced by some Ansatz (like the mean for like apartaments), because we could have very large apartaments and very small ones in the same building, for example.

And since the missing values are only `.4%` of the total values (vide infra), we'll simply discard these missing rows

```python
len(df[df['area'].isnull()])/len(df['area'])
```

```python
df.dropna(subset=['area'], inplace=True)
```

Now let's see the percentage of missing values in each column:

```python
100*df.isnull().sum()/(df.count() + df.isnull().sum())
```

The most alarming missing value here is the condominio. The quartos and banheiros we can drop it -- they're not really expressive.

The anunciante is not really a very important variable, to begin with. So, for now, we'll just leave the missing values from anunciante as they are.

```python
df.dropna(subset=['quartos', 'banheiros'], inplace=True)
100*df.isnull().sum()/(df.count() + df.isnull().sum())
```

Now, to deal with the missing data from condominio we have to note the following: the condominio is set by the building and its related to the services it offers.
Probably a good index for many services in the building is the garage: the more spots there's for a single apartament, probably the steeper the condominio is.

Indeed we see this relationship in the following correlation matrix

```python
df.corr()
```

To fill the missing values from condominio we'll use the K nearest neighboors algorithm based on the garage:

```python
df_nona = df.copy().dropna()
X_train, X_test, y_train, y_test = train_test_split(np.array(df_nona['garagem']).reshape(-1,1), df_nona['condominio'], random_state = 1)
```

```python
knn = KNeighborsRegressor(n_neighbors=10)
knn.fit(X_train, y_train)
predictions_test = knn.predict(X_test)
predictions_train = knn.predict(X_train)
```

Next we'll compare the root mean squared error (RMSE) of the predctions from the train set and the test set to see if we're overfitting.

```python
print(np.sqrt(mean_squared_error(y_test, predictions_test)))
print(np.sqrt(mean_squared_error(y_train, predictions_train)))
print('-'*100)
print(np.std(y_test))
```

So, we're really not overfitting -- if we were the RMSE from the train set would be much smaller than the one from the test set.

At the same time, the RMSE is smaller than the standard deviation, so its a reasonable estimate. We'll go ahead and use this to fill in the `NaN`s

```python
df[df['condominio'].isnull()]['garagem']
```

```python
knn = KNeighborsRegressor(n_neighbors=10)
knn.fit(df_nona[['garagem']], df_nona['condominio'])

X_nan = np.array(df[df['condominio'].isnull()]['garagem']).reshape(-1,1)
predictions_nan = knn.predict(X_nan)
```

Now we'll substitute the missing values for the `predictions_nan`

```python
df[df['condominio'].isnull()]
```

First we create a pandas series from `predictions_nan` with the same index as the null values from `condominio`:

```python
values_nan = pd.Series(predictions_nan, index=df[df['condominio'].isnull()].index)
values_nan
```

And then we fill the missing values:

```python
df['condominio'].fillna(values_nan, inplace=True)
```

```python
df['condominio'].isnull().sum()
```

## Making sense of the data

Now, we'll explore the data itself and see if there's anything wrong with the values it contains.

```python
df.describe()
```

We already see something weird with the `area`: the minimum value if `1` squared meter. The only reasonable assumption would be if that was an storage unit.

```python
df[df['area'] == 1]
```

There are a bunch of apartaments with area 1 but they're not storage units (they have quartos > 1, they have bathrooms and so on). Their price are also quite high.
With that in mind, this can only be an error.

```python
100*len(df[df['area'] == 1])/len(df)
```

Since these represent only `.15%` of the total data its best if we simply drop them.

But first let's check if the second least value for area is reasonable (so that we might have hope of not repeatedly dropping absurd values)

```python
criteria = df['area'] == df[df['area'] != 1]['area'].min()
df[criteria]
```

We see its a very small apartament (3x3 m), but with a reasonable price for that size. It makes sense.

So we'll just drop rows with area == 1:

```python
df = df[df['area'] != 1]
df.describe()
```

What about the maximum values for aluguel, condominio and area?

```python
df[df['aluguel'] == df['aluguel'].max()]
```

```python
df[df['condominio'] == df['condominio'].max()]
```

```python
df[df['area'] == df['area'].max()]
```

The first two are very expensive, but for an initial cleaning we'll leave them as they are.

The row with the maximum area makes less sense.

But we'll leave things as they are, since this is an initial cleaning. A cleaning for analysis and ML (which we do in another notebook) would exclude values above/below some multiple of the Z-score.
